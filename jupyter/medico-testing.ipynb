# Hipotesis, porque no se han contagiado mas los medicos

Si los m√©dicos no han tenido una mayor tasa de contagio, puede ser debido a una inmunidad, aparte de los anticuerpos.

Para ello, calculamos distintos tipos de prevalencia


```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

array_prevalencia = [0.0 ,0.1,0.15,0.2,0.25,0.3,.4,.5,.6]
array_df = []
DIAS=80
CUIDADO=0.993

for prevalencia in array_prevalencia:
    cuidado_real = (994 + (prevalencia*1000) ) / (1000 +( prevalencia*1000 ))
    infectados=1
    array_temp=[]
    for dia in range(1,DIAS):
        infectados=infectados*cuidado_real
        array_temp.append(infectados)
    df_temp = pd.DataFrame(array_temp,columns=(['prevalence '+str(prevalencia)]))
    array_df.append(df_temp)
    print(prevalencia,cuidado_real,array_temp[-1])

array_df
df= pd.concat(array_df, axis=1)
df.plot(legend=True,figsize=(13.5,9), marker='o',title=f"""Porcentaje de medicos infectados al {1- CUIDADO:.3f}% infecci√≥n diaria""")

```
